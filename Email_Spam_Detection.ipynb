Importing Library

```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, mean_squared_error, r2_score
import numpy as np
import kagglehub
import pandas as pd
import string
```

Collecting Dataset

```
path = kagglehub.dataset_download("abdallahwagih/spam-emails")
print("Path to dataset files:", path)
```

```
df = pd.read_csv('../kaggle/input/spam-emails/spam.csv')
print(df.head())
```

Data Preprocessing

```
df.columns = ['label', 'text']
```

```
#Convert Labels to Numbers
#Machine learning models work with numbers, not text labels:

df['label'] = df['label'].map({'ham': 0, 'spam': 1})
```

```
#Clean the Text Data
#Remove punctuation, lowercase everything, and (optionally) remove stopwords:

def clean_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text

df['text'] = df['text'].apply(clean_text)
```

Split the Data

```
#Split your data into training and testing sets to evaluate your model’s performance:

X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.2, random_state=42)
```

Feature Extraction (Vectorization)

```
#Convert text into numbers using CountVectorizer:

vectorizer = CountVectorizer(stop_words='english')
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

Model Training

```
#Train a simple and effective model (Multinomial Naive Bayes):

model = MultinomialNB()
model.fit(X_train_vec, y_train)
```

Model Evaluation

```
# Check how well your model works:

y_pred = model.predict(X_test_vec)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, zero_division=0))
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
print("R2 Score:", r2_score(y_test, y_pred))
```

Predict New Emails

```
def predict_email(text):
    text = clean_text(text)
    vec = vectorizer.transform([text])
    prediction = model.predict(vec)[0]
    return "Spam" if prediction == 1 else "Not Spam"

print(predict_email("Congratulations! You have won a free iPhone. Click here to claim."))
```
